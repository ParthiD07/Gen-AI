```python
import re
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

Text Pre-processing using WordNet lemmatizer

```python
corpus=[]

for i in range(len(email)):
    review=re.sub("[^a-zA-Z]"," ",email["message"][i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review]
    review=" ".join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus,email["message"]) if i<1]
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
# Train Word2vec from scratch
import gensim
model=gensim.models.Word2Vec(words)
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent= [word for word in doc if word in model.wv.index_to_key]
    #print(sent)

    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
# Dependent Features
y= email[list(map(lambda x: len(x)>0,corpus))]
y=pd.get_dummies(y["label"])
y.drop(index=[451,783,2102,2569,3269,3740,3979,4291,4478,4620,5170,5411],inplace=True)
y=y.iloc[:,1].values
```

```python
df = pd.DataFrame()

for i in range(len(X_new)):
    # Create a new DataFrame from X_new[i] and concatenate
    new_row = pd.DataFrame(X_new[i].reshape(1, -1))
    df = pd.concat([df, new_row], ignore_index=True)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

Choosing Random Forest Classifer

```python
from sklearn.ensemble import RandomForestClassifier
classifier= RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```
