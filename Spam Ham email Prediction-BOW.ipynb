```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import SnowballStemmer
stemmer= SnowballStemmer('english')
lemmatizer= WordNetLemmatizer()
```

Text Pre-processing using WordNet lemmatizer

```python
corpus=[]
for i in range(len(email['Message'])):
    review= re.sub('[^a-zA-Z]',' ',email['Message'][i]) # this line substitutes all special characters with blank space
    review= review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

Text Pre-processing using Snowball Stemmer

```python
corpus1=[]
for i in range(len(email['Message'])):
    review= re.sub('[^a-zA-Z]',' ',email['Message'][i]) # this line substitutes all special characters with blank space
    review= review.lower()
    review=review.split()
    review=[stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    review=" ".join(review)
    corpus1.append(review)
```

Building a prediction system

```python
# Function to preprocess the input message
def preprocess_message(message):
    stemmer = SnowballStemmer('english')
    review = re.sub('[^a-zA-Z]', ' ', message)  # Remove special characters
    review = review.lower()  # Convert to lowercase
    review = review.split()  # Split into words
    review = [stemmer.stem(word) for word in review if word not in stopwords.words('english')]
    review = " ".join(review)  # Rejoin into a single string
    return review
# Function to predict spam or ham
def predict_message(message):
    processed_message = preprocess_message(message)
    transformed_message = cv1.transform([processed_message]).toarray()  # Transform using the trained vectorizer
    prediction = model.predict(transformed_message)  # Make prediction using the trained model
    return "Ham" if prediction[0] == 1 else "Spam"
```

```python
# Input and prediction
user_input = input("Enter your message: ")  # Get user input
result = predict_message(user_input)  # Predict
print(f"The message is classified as: {result}")
```
